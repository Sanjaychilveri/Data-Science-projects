IMPORTING DEPENDENCIES

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

LOAD THE DATASET

```python
columns=['sepal_length','sepal_width','petal_length','petal_width','class_labels']
df=pd.read_csv("iris.data",names=columns);
df.head()

```

VISUALIZARION OF DATASET

```python
df.describe()
```

```python
#plot the class labels 
sns.pairplot(df,hue='class_labels');
```

SEPARATING THE I/P AND O/P COLUMNS

```python

data=df.values
X=data[:,0:4]
Y=data[:,4]
```

TRAINING DATA AND TESTING DATA SPLIT

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

MODEL 1: SUPPORT VECTOR MACHINE 

```python
from sklearn.svm import SVC

model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction1= model_svc.predict(X_test);
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(Y_test,prediction1);
print(prediction1)
print(accuracy)
```

MODEL 2: LOGISTIC REGRESSION

```python
from sklearn.linear_model  import LogisticRegression

model_lr=LogisticRegression();
model_lr.fit(X_train,Y_train);
```

```python
prediction2=model_lr.predict(X_test);
print(prediction2)
accuracy2=accuracy_score(Y_test,prediction2);
print(accuracy2)
```

MODEL 3: DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier

model_dt=DecisionTreeClassifier()
model_dt.fit(X_train,Y_train)
```

```python
prediction3=model_dt.predict(X_test)
accuracy3=accuracy_score(Y_test,prediction3)
print(prediction3)
print(accuracy3)
```

```python
#classification report
from sklearn.metrics import classification_report
print("FOR MODEL 1")
print(classification_report(Y_test,prediction1)) 
print("FOR MODEL 2")
print(classification_report(Y_test,prediction2)) 
print("FOR MODEL 3")
print(classification_report(Y_test,prediction3)) 
```
