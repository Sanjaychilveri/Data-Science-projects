```python
import cv2 as cv
image =cv.imread('zoro.jpg')
gray_img=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
```

```python
#invert the gray image to negative image
invert_gray_img=255-gray_img
```

```python
blurred_img=cv.GaussianBlur(invert_gray_img,(21,21),0)
```

```python
#inverting the blurred img
invert_blur_img=255-blurred_img
```

```python
#create the pencil sketch by dividing the grayscale img by the inverted blurred img
pencil_sketch=cv.divide(gray_img,invert_blur_img,scale=256.0)
```

```python
#show the original img and pencil sketch
cv.imshow('orignal image',image)
cv.imshow('pencil sketch',pencil_sketch)
cv.imshow('blurred image',blurred_img)
cv.waitKey(0)
cv.destroyAllWindows()

```

```python
#display the img using matplotlib
import matplotlib.pyplot as plt
fig,axes=plt.subplots(1,3,figsize=(15,5))

axes[0].imshow(cv.cvtColor(image,cv.COLOR_BGR2RGB))
axes[0].set_title('orginal img')
axes[0].axis('off')

axes[1].imshow(pencil_sketch,cmap='gray')
axes[1].set_title('pencil sketch')
axes[1].axis('off')

axes[2].imshow(blurred_img)
axes[2].set_title('blur img')
axes[2].axis('off')

plt.show()
```
